```python
# the data from the last hw
import matplotlib.pyplot as plt
import numpy as np

data = np.loadtxt("circular.txt", float)
plt.imshow(data)
plt.xlabel('Original')
# plt.savefig("original.png")
plt.colorbar()
plt.xlabel(r'$x$ (pixels)')
plt.ylabel(r'$y$ (pixels)')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from cmath import exp, pi

def dft(y):
    '''
    This is a code to do decrete Fourier transform
    '''    
    N = len(y)
    c = np.zeros(N, complex)
    for k in range(N):
        for n in range(N):
            c[k] += y[n]*exp(-2j*pi*k*n/N)
    return c
```

```python

y1 = np.fft.fft2(data)
plt.title("fft")
plt.plot(np.abs(y1[0]))
plt.show()

y2 = dft(data[0])
plt.plot(np.abs(y2))
plt.show()
```

```python
import timeit

# average time for the first loop
%timeit dft(data[0])
```

```python
#total time it took to do all 501loops

start = timeit.default_timer()
i=0
for i in range(501):
    y2 = dft(data[i])
    plt.plot(np.abs(y2))
plt.show()

end = timeit.default_timer()

print(end - start, " Seconds taken to finish calculation")
```

```python
# vs the amount of time to do everything 

start = timeit.default_timer()

y1 = np.fft.fft2(data)
plt.title("fft")
plt.plot(np.abs(y1))
plt.show()

end = timeit.default_timer()
print(end - start, " Seconds taken to finish calculation")
```

As we can see numpy's fast fourier transform can perform the calculation of the entire 501 array faster than it takes our regular algorithm to do one calculation.
