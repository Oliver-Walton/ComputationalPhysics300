### 23.5 One Rule Algorithm

In the above section, we have developed a simple classifier based on the graph analysis, which could almost reach 100 % of success rate. However, this is not machine learning! What we really want is to let computer find the rule automatically. 
So we need a more automtated algorithm.

#### 23.5.1 Discetization

Unlike human, computers like discrete number!
It is usally useful to transform the data from continuous numbers to the discrete numbers. The simplest way is to 
take the average value, set all the values which are lower than the mean as 0, and all the values higher than mean as 1. This could be simply done by one line with the help of numpy.

After we have the table, we can apply "One Rule" algorithm based on the frequency, which a simple, yet accurate, classification algorithm based on one rule. 

The algorithm is the following. 
- 1, For each predictor,
     - 1.1 For each value of that predictor, make a rule as follows;
           - 1.1.1 Count how often each value of target (class) appears
           - 1.1.2 Find the most frequent class
           - 1.1.3 Make the rule assign that class to this value of the predictor
     - 1.2 Calculate the total error of the rules of each predictor
- 2, Choose the predictor with the smallest total error.
